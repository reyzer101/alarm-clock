```python
import winsound
import time
import os

def sound():
    for _ in range(9): 
        winsound.MessageBeep(-1)
    time.sleep(2) 

def alarm(n):
    print("\nWait time:", n, "seconds.")
    time.sleep(n)
    sound()

def get_integer_input(prompt):
    while True:
        try:
            value = int(input(prompt))
            if value < 0:
                raise ValueError("Input must be a positive integer")
            return value
        except ValueError as e:
            print("Error:", e)

def set_alarm():
    print("What unit of time do you want to wait?\n"
          " (1) Hours\n"
          " (2) Minutes\n"
          " (3) Seconds\n"
          " (4) Combination")
    
    choice = input(": ")
    
    if choice == '1':
        hours = get_integer_input("Enter the desired hours: ")
        wait_time = hours * 3600
    elif choice == '2':
        minutes = get_integer_input("Enter the desired minutes: ")
        wait_time = minutes * 60
    elif choice == '3':
        seconds = get_integer_input("Enter the desired seconds: ")
        wait_time = seconds
    elif choice == '4':
        hours = get_integer_input("Hours: ")
        minutes = get_integer_input("Minutes: ")
        seconds = get_integer_input("Seconds: ")
        wait_time = (hours * 3600) + (minutes * 60) + seconds
    else:
        print("Invalid choice")
        return

    alarm(wait_time)

def main():
    print("Welcome to the Alarm Program!")
    while True:
        set_alarm()
        again = input("Do you want to set another alarm? (yes/no): ")
        if again.lower() != 'yes':
            break
    print("Exiting...")

if __name__ == "__main__":
    main()
```
